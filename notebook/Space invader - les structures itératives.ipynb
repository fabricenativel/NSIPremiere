```python
dessine_pixel(2, 5, couleur)

```

On constate que seul le chiffre des abcisses est modifi  chaque ligne. Notons le **k**, chaque ligne est en fait :

```python
dessine_pixel(k,5)
```
avec k qui vaut succesivement 2 puis 3 puis... jusqu' 8 (inclus).

On peut remplacer les 8 lignes par seulement deux lignes en faisant une boucle :
```python
for k in [2,3,4,5,6,7,8] :
    dessine_pixel(k,8)
```



<div style="margin:2px;border:1px solid black;padding:10px;">

+ Une structure itrative ou boucle permet de rpter un ensemble d'instructions.
```python
    # cette boucle rpte deux fois les instructions
    for _ in range(2):
        instructions
```
+ On utilise une collection de valeurs pour indiquer le nombre d'itrations (ou passage dans la boucle) :
```python
    # cette boucle rpte trois fois les instructions
    for i in [1, 2, 3]:
        instructions
```
+ Dans ce cas, la variable ``i`` changera de valeur  chaque itration et prendra tour  tour les valeurs de la collection.
</div>

On peut donc crer une structure itrative pour chaque lignes, et notre code devient :

<p>Si on considre la seconde ligne de mon space invader&nbsp;:</p>

```python

```

Il y a deux mthodes principales pour dessiner cette mme ligne 15 pixels sur la droite&nbsp;:

+ ajouter 15  toutes les valeurs de la liste des valeurs de k

```python
for k in [17,18,19,20,21] :

```
+ ajouter directement 15  k dans l&#8217;instruction qui permet de dessiner un pixel

```python 
for k in [2,3,4,5,6] :
```
